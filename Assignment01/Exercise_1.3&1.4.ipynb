```python
def load_text(filepath='data.txt'):
    """
    Load text from a file.

    Args:
        filepath (str): Path to the file to be loaded.

    Returns:
        str: The content of the file as a string.
    """
    with open(filepath, "r", encoding="utf-8") as file:
        content = file.read()
    return content

def preprocess(text):
    """
    Preprocess the input text by lowercasing and removing punctuation.

    Args:
        text (str): The input text to preprocess.

    Returns:
        list: A list of tokens (words) after preprocessing.
    """
    text = text.lower()
    for ch in ['–', '—', '‘', '’', '“', '”', '…', '•', '–', '—']:
        text = text.replace(ch, ' ')

    text = text.translate(str.maketrans('', '', string.punctuation))

    tokens = text.split()

    # Keep only alphabetic tokens and filter short words (optional)
    clean_tokens = []
    for word in tokens:
        if word.isalpha() and len(word) > 1:
            clean_tokens.append(word)
    return clean_tokens

def analyse(tokens):
    """
    Analyze the tokens to compute unigrams, bigrams, and trigrams, and their frequencies.

    Args:
        tokens (list): A list of tokens (words).

    Returns:
        tuple: A tuple containing frequency dictionaries for unigrams, bigrams, and trigrams.
    """
    unigrams = {}
    bigrams = {}
    trigrams = {}

    for ch in tokens:
      unigrams[ch] = unigrams.get(ch, 0) + 1
    for i in range(len(tokens) - 1):
      bg = tuple(tokens[i:i+2])
      bigrams[bg] = bigrams.get(bg, 0) + 1

    for i in range(len(tokens) - 2):
      tg = tuple(tokens[i:i+3])
      trigrams[tg] = trigrams.get(tg, 0) + 1

    n = 30
    # We choose to display only 30 due to the complexity and lack of readability given the number of unique tokens.
    top_unigrams = sorted(unigrams.items(), key=lambda item: item[1], reverse=True)[:n]
    top_bigrams = sorted(bigrams.items(), key=lambda item: item[1], reverse=True)[:n]
    top_trigrams = sorted(trigrams.items(), key=lambda item: item[1], reverse=True)[:n]

    print("Top 30 Unigrams:", top_unigrams)
    print("Top 30 Bigrams:", top_bigrams)
    print("Top 30 Trigrams:", top_trigrams)


    return unigrams, bigrams, trigrams

def hist_plot(freq_dict, title="Histogram"):
    """
    Plot a histogram for the given frequency dictionary.

    Args:
        freq_dict (dict): A dictionary containing items and their frequencies.
        title (str): The title of the histogram.
    """
    # We choose to display only 30 due to the complexity and lack of readability given the number of unique tokens.
    n = 30
    sorted_words = sorted(freq_dict.items(), key=lambda item: item[1], reverse=True)[:n]
    items = [str(item) for item, count in sorted_words]
    frequencies = [count for item, count in sorted_words]


    plt.figure(figsize=(10, 6))
    plt.bar(items, frequencies, color='skyblue', edgecolor='black')
    plt.title(title)
    plt.xlabel('Items')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
